```python
# Importing packages
import pandas as pd
import os
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', None)
import os
from natsort import natsorted
```

```python
# Run through each csv file in the data directory and remove the id/ string from the record_id column

path_data = 'data_files/data'
dir_list = os.listdir(path_data)
print(dir_list)

for file in dir_list:
    if '.DS' in file:
        pass
    
    else:
        
        try:
            directory = path_data+file
            df_data = pd.read_csv(directory, index_col=0)
            df_data['record_id'] = df_data['record_id'].str.replace('id/','')
            df_data.to_csv(directory, index=False)

        except Exception as e:
            print(directory)
            print(e)
```

```python
# Creating a function that goes through each CSV file in the data_csv_content directory and
# creates a column with the entry of its record name

path_csv = 'data_files/data_csv_content'
dir_list = os.listdir(path_csv)
print(dir_list)

for file in dir_list:
    if '.DS' in file:
        pass
    
    else:
        
        try:
            directory = path_csv+file
            df_ref = pd.read_csv(directory, index_col=0)
            df_ref['record_id'] = file[:-4]
            df_ref.to_csv(directory, index=False)

        except Exception as e:
            print(directory)
            print(e)
```

```python
path_data = 'data_files/data'
dir_list = natsorted(os.listdir(path_data))
print(dir_list)

# Combining all the main data
df_append = pd.DataFrame()

for file in dir_list:
    if '.DS' in file:
        pass
    
    else:

    #append all files together
    
        main_direct = path_data+'/'+file    
        df_temp = pd.read_csv(main_direct, index_col=False)
        df_append = pd.concat([df_append,df_temp], ignore_index=True)

df_append.to_csv('data_files/data_concat/combined_main_data_files.csv', sep=";", index=False)
```

```python
# Time to perform some data cleaning for the labels
```

```python
overall = pd.read_csv("data_files/data_concat/combined_main_data_files.csv", sep=";")

filtered_df = overall[overall["Images"]=="Yes"]

filtered_df.head()
```

```python
# While the record occasionally contains the name of the emperor, sometimes a coin is attributed to an emperor in record_description
# but there is no image of the emperor. This renders the record useles. 
# Usually in the Obverse column, if the image of an emperor is visible, it is denoted with "Head of" or "Bust"
# Replacing "Bust" in text to "Head"
filtered_df['Obverse'] = filtered_df['Obverse'].str.replace('Bust', 'Head')
```

```python
# Creating new columns for the spliting of the Obverse string
filtered_df[['cola','colb', 'colc']] = filtered_df.Obverse.str.split(":", expand =True)
```

```python
filtered_df.head()
```

```python
# For our purposes, column colb is of interest
# Let us separate it down further by spliting according to ',' to get the first item
filtered_df['colab'] = filtered_df.colb.str.split(',', expand = True)[0]
filtered_df.head()
```

```python
# Removing the 'Head of ' label and renaming the column 'Emperor'
filtered_df.rename(columns={'colab': 'Emperor'}, inplace=True)
filtered_df["Emperor"] = filtered_df["Emperor"].str.replace("Head of ", "")
```

```python
filtered_df['Emperor'].value_counts()
```

```python
# Here we will need to do some manual cleaning in Excel
# Let us first export this overview to a csv file
emperor_counts = filtered_df['Emperor'].value_counts()
emperor_counts_df = emperor_counts.reset_index()
emperor_counts_df.columns = ['Emperor', 'Count']
emperor_counts_df.to_csv('data_files/data_concat/emperor_list.csv', sep=';', index=False) 
```

```python
# Open emperor list

emperor_list = pd.read_csv('data_files/data_concat/emperor_list.csv', sep=';')
emperor_list.head()
```

```python
# Perform pd merge

filtered_df = pd.merge(filtered_df, emperor_list, on='Emperor', how='left')
filtered_df.head()
```

```python
# We also have entries in cola that did not produce entries in colb for processing
cola_items = filtered_df[filtered_df['colb'].isna()]
cola_items.head(20)
```

```python
# Performing the same procedure for cola
filtered_df['colaa'] = filtered_df.loc[filtered_df['colb'].isna(), 'cola'].str.split(',', expand=True)[0]
filtered_df[filtered_df['colb'].isna()].head()
```

```python
# Removing the 'Head of ' label
filtered_df["colaa"] = filtered_df["colaa"].str.replace("Head of ", "")
```

```python
colaa_counts = filtered_df['colaa'].value_counts()
```

```python
emperor_counts_df_cola = colaa_counts.reset_index()
emperor_counts_df_cola.columns = ['Emperor', 'Count']
emperor_counts_df_cola.to_csv('data_files/data_concat/emperor_list_cola.csv', sep=';',index=False) 
```

```python
# Doing the emperor categorisation in that file then merging it to the overall dataframe
colaa_emps = pd.read_csv('data_files/data_concat/emperor_list_cola.csv', sep=';')
colaa_emps.head(5)
```

```python
# Perform a left merge on 'overall' and 'colaa_emps' based on 'cola' and 'Emperor'
filtered_df = pd.merge(filtered_df, colaa_emps, how='left', left_on='colaa', right_on='Emperor')

# Fill NaN values in 'emperor_categorisation_x' with values from 'emperor_categorisation_y'
filtered_df['emperor_categorisation_x'].fillna(filtered_df['emperor_categorisation_y'], inplace=True)

# Drop redundant columns
filtered_df.drop(['emperor_categorisation_y'], axis=1, inplace=True)

# Rename the filled 'emperor_categorisation_x' column
filtered_df.rename(columns={'emperor_categorisation_x': 'emperor_categorisation'}, inplace=True)
```
