### min_df apply gray, normalize, rotate, color jitter

```python
# Rotating images

path = "data_files/resized_images"

#done_changes = ['gray', 'normalized']


for folder in min_df:

    path_folder = path + '/' + folder
    
    for image in os.listdir(path_folder):

        if any([x in image for x in min_df_1]):

            try: 

                image_path = path_folder + '/' + image
                print(image_path)
                
                orig_img = Image.open(Path(image_path))

                transform = T.RandomRotation(degrees=(0,180))

                for i in range(8):

                    new_image = transform(orig_img)                       
                
                    # Save the image
                    save_path = image_path +f"_rotated_{i}.jpg"

                    new_image.save (save_path)


            except Exception as e:
                    print(e)

        elif any([x in image for x in min_df_2]):
            
            try:
                image_path = path_folder + '/' + image
                print(image_path)
                
                orig_img = Image.open(Path(image_path))

                transform = T.RandomRotation(degrees=(0,180))

                for i in range(4):

                    new_image = transform(orig_img)                       
                
                    # Save the image
                    save_path = image_path +f"_rotated_{i}.jpg"

                    new_image.save (save_path)


            except Exception as e:
                    print(e)

        elif any([x in image for x in min_df_3]):
            
            try:
                image_path = path_folder + '/' + image
                print(image_path)
                
                orig_img = Image.open(Path(image_path))

                transform = T.RandomRotation(degrees=(0,180))

                for i in range(2):

                    new_image = transform(orig_img)                       
                
                    # Save the image
                    save_path = image_path +f"_rotated_{i}.jpg"

                    new_image.save (save_path)


            except Exception as e:
                    print(e)




        else:
             pass
```

```python
path = "data_files/resized_images"

# Gray Scale

for folder in min_df:

    path_folder = path + '/' + folder
    
    for image in os.listdir(path_folder):
        
        try:
            image_path = path_folder + '/' + image
            
            orig_img = Image.open(Path(image_path))

            gray_img = T.Grayscale()(orig_img)

            # Save the image
            save_path = image_path +'_gray.jpg'

            gray_img.save (save_path)
        
        except Exception as e:
            print(e)
            pass


```

```python
# The Min Size has doubled.
# Now let us normalize the images

# Normalize

path = "data_files/resized_images"

#done_changes = ['gray']

for folder in min_df:

    path_folder = path + '/' + folder
    
    for image in os.listdir(path_folder):

        #if any([x in image for x in done_changes]):

            #pass
        
        #else: 
                try:
            
                    image_path = path_folder + '/' + image
                    
                    
                    orig_img = Image.open(Path(image_path))                
            
                    # define custom transform function
                    transform = T.Compose([T.ToTensor()])

                    # transform the pIL image to tensor 
                    # image
                    img_tr = transform(orig_img)
                    
                    # Convert tensor image to numpy array
                    img_np = np.array(img_tr)
                    

                    # calculate mean and std
                    #mean, std = img_tr.mean([1,2]), img_tr.std([1,2])
                    
                    ##################################################
                    # Fr Transfer Learning ResNet 101 we need to update the normalisation to using the mean and standard deviation of the training set.
                    ##################################################

                    # From the following website: https://pytorch.org/vision/main/models/generated/torchvision.models.resnet101.html
                    mean = [0.485, 0.456, 0.406]
                    std = [0.229, 0.224, 0.225]


                    # define custom transform
                    # here we are using our calculated
                    # mean & std
                    transform_norm = T.Compose([T.ToTensor(), T.Normalize(mean, std)])               

                    
                    # get normalized image
                    normalized_img = transform_norm(orig_img)

                    # convert this image to numpy array
                    #normalized_img = np.array(normalized_img)


                    # convert back to image
                    transform = T.ToPILImage()

                    normalized_img = transform(normalized_img)
                                                    
                                          
                    
                    # Save the image
                    save_path = image_path +'_normalized.jpg'

                    normalized_img.save (save_path)

                    

                except Exception as e:
                     print(e)
            
```

```python
path = "data_files/resized_images"

# Random Flip

for folder in min_df:

    path_folder = path + '/' + folder
    
    for image in os.listdir(path_folder):
        
        try:
            image_path = path_folder + '/' + image
            
            orig_img = Image.open(Path(image_path))

            gray_img = T.RandomVerticalFlip(p=1)(orig_img)

            # Save the image
            save_path = image_path +'_vertiflip.jpg'

            gray_img.save (save_path)
        
        except Exception as e:
            print(e)
            pass
```

```python
# Now let us normalize the images

# Normalize

path = "data_files/resized_images"

done_changes = ['gray']

for folder in lower_quart:

    path_folder = path + '/' + folder
    
    for image in os.listdir(path_folder):

        if any([x in image for x in done_changes]):

            pass
        
        else: 
                try:
            
                    image_path = path_folder + '/' + image                 
                    
                    orig_img = Image.open(Path(image_path))                
            
                    # define custom transform function
                    transform = T.Compose([T.ToTensor()])

                    # transform the pIL image to tensor 
                    # image
                    img_tr = transform(orig_img)
                    
                    # Convert tensor image to numpy array
                    img_np = np.array(img_tr)

                    # calculate mean and std
                    #mean, std = img_tr.mean([1,2]), img_tr.std([1,2])

                    #################################################
                    mean = [0.485, 0.456, 0.406]
                    std = [0.229, 0.224, 0.225]
                    #################################################
                    

                    # define custom transform
                    # here we are using our calculated
                    # mean & std
                    transform_norm = T.Compose([T.ToTensor(), T.Normalize(mean, std)])               

                    
                    # get normalized image
                    normalized_img = transform_norm(orig_img)

                    # convert this image to numpy array
                    #normalized_img = np.array(normalized_img)


                    # convert back to image
                    transform = T.ToPILImage()

                    normalized_img = transform(normalized_img)                              
                    
                    # Save the image
                    save_path = image_path +'_normalized.jpg'

                    normalized_img.save (save_path)

                except Exception as e:
                     print(e)
            
```

```python
# Color Jitter

path = "data_files/resized_images"

done_changes = ['gray', 'normalize', 'rotated']

for folder in lower_quart:

    path_folder = path + '/' + folder
    
    for image in os.listdir(path_folder):

        if any([x in image for x in done_changes]):
             pass
        
        else:

            try:
        
                image_path = path_folder + '/' + image
                print(image_path)
                
                orig_img = Image.open(Path(image_path))

                transform = T.ColorJitter(brightness=1.0, contrast=0.5, saturation=1, hue=0.1)

                new_image = transform(orig_img)

                # Save the image
                save_path = image_path +"_colorjitter.jpg"

                new_image.save (save_path)


            except Exception as e:
                    print(e)
```
